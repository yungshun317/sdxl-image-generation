```python
import torch
from diffusers import StableDiffusionXLPipeline

pipe = StableDiffusionXLPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16
)
pipe = pipe.to("cuda")

prompt = "a photo of an astronaut riding a horse on mars"
image = pipe(prompt).images[0]
image.save("astronaut.png")
```

```python
import matplotlib.pyplot as plt

img = plt.imread('astronaut.png')
plt.imshow(img)
```

# Unconditional Image Generation

```python
from diffusers import DiffusionPipeline

generator = DiffusionPipeline.from_pretrained(
    'anton-l/ddpm-butterflies-128', use_safetensors=True
)
generator.to('cuda')

image = generator().images[0]
image.save('generated_image.png')

img = plt.imread('generated_image.png')
plt.imshow(img)
```

```python
# Seed
pipeline = DiffusionPipeline.from_pretrained(
    'anton-l/ddpm-butterflies-128', use_safetensors=True
)
pipeline = pipeline.to('cuda')

generator = torch.Generator('cuda').manual_seed(0)

image = pipeline(generator=generator).images[0]
image.save('generated_image.png')

img = plt.imread('generated_image.png')
plt.imshow(img)
```

```python
# num_inference_steps
pipeline = DiffusionPipeline.from_pretrained(
    'anton-l/ddpm-butterflies-128', use_safetensors=True
)
pipeline = pipeline.to('cuda')

generator = torch.Generator('cuda').manual_seed(0)

image = pipeline(generator=generator, num_inference_steps=100).images[0]
image.save('generated_image.png')

img = plt.imread('generated_image.png')
plt.imshow(img)
```
